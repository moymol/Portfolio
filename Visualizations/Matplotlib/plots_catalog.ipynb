A catalog of several plots, and formats, is found in this notebook. The main idea is to have the right code for the right plot, and do not waste time by coding the plots again and again.

```python
import numpy as np
import matplotlib.pyplot as plt
```

The format is builded by considering design elements that improve the visualization of the plot

```python
def plot_format(l_title,xlabel,ylabel,fontsize,spines):
#l_title defines the label for title,
#x_label is for the x-axis label, likewise for y_label.
#fontsize is numeric;
#and "spines" is an array which defines what axis are shown
#initializes the plot and subplots
    fig,ax = plt.subplots(figsize=(7,5))
    
#format for title and axis labels
    ax.set_title(label=l_title,
                 loc='left',
                 pad=40,
                 fontsize=fontsize*1.5,
                 color='#545454')
    ax.set_xlabel(xlabel='x label',
                  labelpad=10,
                  fontsize=fontsize,
                  color='#B1B1B1')
    ax.set_ylabel(ylabel='y label',
                  labelpad=10,
                  fontsize=fontsize,
                  color='#B1B1B1')
    
#Format for the ticks and spines of the chart
    ax.spines['right'].set_visible(spines[0])
    ax.spines['top'].set_visible(spines[1])
    ax.spines['left'].set_visible(spines[2])
    ax.spines['bottom'].set_visible(spines[3])
    
    ax.spines['left'].set_color('#B1B1B1')
    ax.spines['bottom'].set_color('#B1B1B1')
    # Only show ticks on the left and bottom spines
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.tick_params(direction='out', length=6, width=1,
                   colors='#B1B1B1',labelsize=fontsize)
    return(fig,ax)
```

```python
#Default format variables
l_title='Descriptive title for the plot'
xlabel='x label'
ylabel='y label'
fontsize=22
spines=[0,0,1,1]
```

## Scatter plot

Do you want to see the relationship among two (or more) variables? This is the chart you are looking for.

```python
#Scatter plot
def scatter_plot(x_,y_,color_points,scatter_size):
    fig,ax=plot_format(l_title,xlabel,ylabel,fontsize,spines)
    ax.scatter(x_,y_,color=color_points,s=scatter_size,alpha=.4)
```

### Examples
#### 2D functions

```python
#Creating the dataset
x_ = np.linspace(0,200,200)
y_ = 0.25*x_ -1

#plot the dataset
scatter_plot(x_,y_,'#3C77B5',40)
plt.show()
```

#### Clusters

```python
#Creating the dataset
x_1=np.random.uniform(1,3,50).tolist()
x_2=np.random.uniform(5,7,50).tolist()
x_3=np.random.uniform(8,10,50).tolist()
x_ =np.append(np.append(x_1,x_2),x_3)
y_ =np.append(np.append(x_2,x_3),x_1)

#plot the dataset
scatter_plot(x_,y_,'#3C77B5',90)
plt.show()
```

#### bubble plot

```python
#Creating the dataset
x_=np.random.uniform(-1,1,70)
y_=x_*x_/2 - np.power(x_,4)/4

s_=np.zeros(len(y_))
for i in range(len(s_)):
    if x_[i] <= 0.0:
        s_[i] = 10
    elif x_[i] > 0.0 and x_[i]<= 0.6:
        s_[i] = 120
    else:
        s_[i] = 900

#plotting the bubbles
scatter_plot(x_,y_,'#3C77B5',s_)
```
